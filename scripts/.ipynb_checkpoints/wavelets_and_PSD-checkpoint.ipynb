```python
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
#from scipy import signal
import pandas as pd
import pywt
import matplotlib.dates as mdates
```

### functions

```python
def create_white_noise(n: int, period:float, amp: float, phase=None):
    from numpy import cos, pi
    """
    n: vecor length
    period: time
    amp: signal amplitude
    phase(optional): displacement of the wave with respect to the origin 
    """
    if phase:
        wave =[amp * cos((i*pi)/period + phase) for i in range(n)]
        wave = np.array(wave)
        return wave
    else:
        wave =[amp * cos((i*pi)/period) for i in range(n)]
        wave = np.array(wave)
        return wave
```

```python
def add_noise(x:pd.Series, noise:np.asarray)->np.asarray:
    x_no_nan = x.copy()
    nan_idx = x_no_nan.isna()
    x_no_nan.loc[nan_idx] = 0
    
    r = x_no_nan + noise
    r = np.array(x_no_nan)
    return r
```

```python
# Ruidos
noise1 = create_white_noise(n=3500, period=1_000, amp=0.01, phase=0.25*np.pi)
#noise2 = create_white_noise(n=10000, period=1_000, amp=0.01, phase=0)
```

## wavelets analysis



```python
for col, depth in zip([1, 4, 10, 14, 19],['8','16','32','45','53']):
    r = add_noise(data.iloc[0:3500, col], noise1)
    # Define the wavelet function
    wavelet = 'morl'

    # Perform the continuous wavelet transform
    coeffs, freqs = pywt.cwt(r, np.arange(1, 36), wavelet)
    # Plot the scaleogram
    plt.figure(figsize=(8, 6))
    plt.imshow(np.abs(coeffs), extent=[1, len(r), 1/freqs[-1], 1/freqs[0]],
               cmap='jet', aspect='auto', interpolation='nearest')
    plt.colorbar()
    plt.title(f'Wavelet Scaleogram of ADCP at: {depth}m')
    plt.xlabel('Time (days)')
    plt.ylabel('T(hours)')
    plt.show()
```
